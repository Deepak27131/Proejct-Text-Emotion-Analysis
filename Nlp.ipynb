```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
df = pd.read_csv('train.txt', sep = ';',header = None, names=['text', 'emotion'])
df
```

```python
unique_emotion  = df['emotion'].unique()
emotion_number = {}

i = 0
for emo in unique_emotion:
    emotion_number[emo] = i
    i += 1
    
df['emotion'] = df['emotion'].map(emotion_number)
```

```python
df
```

```python
# converted all text intop lowercase

df['text'] = df['text'].apply(lambda x : x.lower())
```

```python
df
```

```python
# Removing the panctuation because it destract the model it no meaning the panctuation

import string 

def remove_punc(text):
    return text.translate(str.maketrans('','',string.punctuation))
```

```python
df['text'] = df['text'].apply(remove_punc)
df
```

```python
# remove number in the text data because it no meaning in the text columns 

def remove_num(text):
    new = ""
    for i in text:
        if not i.isdigit():
            new = new + i
    return new

df['text'] = df['text'].apply(remove_num)

df
```

```python
# removeing emoji in dataset then remove it

def remove_emoji(text):
    new = ""
    for i in text:
        if i.isascii():
            new = new + i
    return new

df['text'] = df['text'].apply(remove_emoji)

df
```

```python
# Stopword removeing :It important thing -- it removereing repeated value like (for ,in,is,or, the, of ,and,it,etc) 

import  nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# checking stopword in englsih word 
stop_word = set(stopwords.words('english'))
stop_word
```

```python
df.loc[1]['text']
```

```python
# removing stop word
def remove(text):
    words = text.split()
    cleaned = []
    for i in words:
        if not i in stop_word:
            cleaned.append(i)
        
    return ' '.join(cleaned)
```

```python
df['text'] = df['text'].apply(remove)
df
```

# Using Bag of word Naive Bay 

```python
from sklearn.feature_extraction.text import CountVectorizer , TfidfVectorizer

bow_vectorizer = CountVectorizer()
x_train_bow = bow_vectorizer.fit_transform(X_train)
x_text_bow = bow_vectorizer.transform(X_test)
```

```python
x_train_bow
```

```python
x_text_bow
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

nb_model = MultinomialNB()
nb_model.fit(x_train_bow,y_train)
```

```python
y_pred_nb = nb_model.predict(x_text_bow)
accuracy_nd = accuracy_score(y_pred_nb,y_test)*100
print("Accuracy of Naive Baye using BOW :",accuracy_nd)
```

# Using TF -IDF Logistic Regression

```python

tf_vectorizer = TfidfVectorizer()
x_train_tf = tf_vectorizer.fit_transform(X_train)
x_text_tf = tf_vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

tf_model = LogisticRegression()
tf_model.fit(x_train_tf,y_train)
```

```python
y_pred_tf = tf_model.predict(x_text_bow)
accuracy_tf = accuracy_score(y_pred_tf,y_test)*100
print("Accuracy of Naive Baye using BOW :",accuracy_tf)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier

vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

lr = LogisticRegression()
nb = MultinomialNB()
svm = SVC(probability=True)

voting = VotingClassifier(
    estimators=[
        ('lr', lr),
        ('nb', nb),
        ('svm', svm)
    ],
    voting='soft'
)

voting.fit(X_train_vec, y_train)
y_pred = voting.predict(X_test_vec)
```

```python
accuracy_score(y_pred,y_test)*100
```

```python
# voting.fit(X_train_vec, y_train)

# 🔹 Accuracy check
models = {
    "Logistic Regression": lr,
    "Naive Bayes": nb,
    "SVM": svm,
    "Voting Ensemble": voting
}

for name, model in models.items():
    model.fit(X_train_vec,y_train)
    y_pred = model.predict(X_test_vec)
    print(name, "accuracy:", accuracy_score(y_test, y_pred))
```
